```python
import pandas as pd
# 1. Load the dataset
df=pd.read_csv('/content/Titanic-Dataset.csv')
print(df)
```

```python
# 2. Understand the shape and dimension of the data
print("---- Shape and Dimension ----")
print(f"Dataset Shape:{df.shape}")
print(f"Dataset Dimension:{df.ndim}")

```

```python
# 3. Use .info() to understand data structure and types
print("---- Data Info ----")
print(df.info())
print("\n")
```

```python
# 4. Use .describe() for statistical summaries
print("---- Descriptive Statistics (Numerical) ----")
print(df.describe())
print("\n")

print("---- Descriptive Statistics (Categorical) ----")
print(df.describe(include=['object']))
print("\n")
```

```python
# 5. Identify Numerical vs Categorical variables & document data types

print("---- Numerical Variables ----")
numerical_variables=df.select_dtypes(include=['int64','float64'])
print(numerical_variables.columns.tolist())
print("\n---- Categorical Variables ----")
categorical_variables = df.select_dtypes(include=['object'])
print(categorical_variables.columns.tolist())
print("\n")
print("---- Data Types ----")
print(df.dtypes)
```

```python
# 6. Check for the missing values in dataset
print("---- Missing Values Count ----")
print(df.isnull().sum())
print("\n")
```

```python
# 7. Identifying unique values in categorical columns
print("---- Unique Values in Categorical Columns ----")
categorical_column=df.select_dtypes(include=['object'])
for col in categorical_column:
    print(df[col].value_counts().head(5))
    print("\n")
```
